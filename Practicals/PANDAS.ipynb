```python
def total_vaccinations_world(vac):
    sum = 0.0   
    l = lambda x: x
    for value in vac:
        sum = sum+l(value)
    return sum

df = df.dropna()
ans = total_vaccinations_world(df.total_vaccinations)
print(f'total vaccinated people all over the world = {ans}')
```
